```python
import numpy as np
import pandas as pd

sheet_name = 'CONSOLIDATED STATEMENTS OF OPER'
file_path = '/Users/evan/Desktop/chatbot/storage and retreival testing grounds/MVST_2022_FY_.xlsx'
xls = pd.ExcelFile(file_path)
df = pd.read_excel(xls, sheet_name=sheet_name)
print(df)

target_column = 'Target'
feature_columns = ['feature1', 'feature2']

x = df[feature_columns].values
y = df[target_column].values

def gradient_descent(x, y, learning_rate=0.01, num_iterations=1000):
    m, n = x.shape
    theta = np.zeros(n)
    price_history = []

    for i in range(num_iterations):
        predictions = np.dot(x, theta)
        error = predictions - y
        gradient = (1/m) * np.dot(x.T, error)
        theta -= learning_rate * gradient

        price = (1/(2*m)) * np.sum(error ** 2)
        price_history.append(price)

    return theta, price_history

# Usage
optimal_theta, price_history = gradient_descent(x, y)
print("Optimal Theta:", optimal_theta)
print("Future Price:", price_history[-1])
```

```python
import pandas as pd

def convert_dates(file_path, sheet_name):
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    year_row_index = 0

    for col in df.columns[1:]:
        try:
            year = col.split()[-1]
            date_string = f"12/31/{year}"
            df.iat[year_row_index, df.columns.get_loc(col)] = pd.to_datetime(date_string, format='%m/%d/%Y')
        except (ValueError, IndexError):
            # Handle columns with unexpected names
            print(f"Skipping column with unexpected name: {col}")
            continue

    output_file_path = 'output_excel_file40.xlsx'
    df.to_excel(output_file_path, index=False)

    print(f"Excel file saved with year rows converted to the last day of December: {output_file_path}")

file_path = '/Users/evan/Desktop/chatbot/storage and retreival testing grounds/MVST_2022_FY_.xlsx'
sheet_name = "CONSOLIDATED STATEMENTS OF OPER"
convert_dates(file_path, sheet_name)
```

```python
import pandas as pd
def growth():
    excel_file = '/Users/evan/Desktop/chatbot/DataPriming3/MVST_2022_FY_RUNINNG.xlsx'
    sheet_name = 'CONSOLIDATED STATEMENTS OF OPER'
    df_data = pd.read_excel(excel_file, sheet_name, index_col=0, na_values= "")

    df_data = df_data.replace({'\$': '', ',': ''}, regex=True).apply(pd.to_numeric, errors='coerce')

    for col in df_data.columns:
        prev_col = df_data[col].shift(1)
        growth = (((df_data[col] - prev_col) / prev_col) * 100).round(2)
        df_data[col + ' Growth'] = growth

    column_order = []
    for col in df_data.columns:
        if " Growth" in col:
            continue
        column_order.append(col)
        column_order.append(col + ' Growth')
    df_data = df_data[column_order]

    growth_columns = [col for col in df_data.columns if " Growth" in col]
    df_data[growth_columns] = df_data[growth_columns].applymap(lambda x: f"{x:.2f}%")

    output_excel_file = 'output_file69.xlsx'
    df_data.to_excel(output_excel_file)

    print(df_data)


growth()



```

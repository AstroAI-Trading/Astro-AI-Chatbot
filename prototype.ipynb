Next, we should create a garbage collecting function. This function will allow us to filter any data we wish not to be in the cache.

```python
# Set this to false if you do not want garbage collecting
collect_garbage = True

def garbage_collector(vars__):
    if collect_garbage:
        for v in vars__:
            if v in globals():
                del globals()[v]
        collected = gc.collect()
        print(f'This is what was collected {collected}.')
    else:
        pass

print("It's good")
```

I will show you two methods on how to use the pandas library to read data. The first method involves pandas reading an excel file. 

```python
MVST_df = pd.ExcelFile('MVST_2022_FY_.xlsx')
# Use the MVST_df variable to open certain pages in the excel document.
# Using a dictionary as pictured below can be useful to organizing the code.
pages = {
    'Cover': pd.read_excel(MVST_df, 'Cover'),
    'Audit': pd.read_excel(MVST_df, 'Audit Information'),
    'Consolidated': pd.read_excel(MVST_df, 'CONSOLIDATED BALANCE SHEETS')
}

'''
Be aware these functions use strings to map the file and the pages.
Make sure these strings are spelled correctly. They are case-sensitive.
'''

'''
Let us create a Boolean expression to check if this file exists.
It is a good idea to make sure your program is always checking for any possible errors.
'''

if os.path.isfile('MVST_2022_FY_.xlsx'):
    display(pages['Cover'], pages['Audit'], pages['Consolidated'])
else:
    raise Exception('Excel file not found. Please make sure it is set to the correct path.')
```

Using the matplotlib library provided by Python, we can show the data using a bar graph

```python
x = []
y = []

with open('Apple_Financial_Statement.csv', 'r') as apple_csv:
    lines = csv.reader(apple_csv, delimiter=',')

    for rows in lines:
        x.append(rows[0])
        y.append(rows[2])
    
    plt.plot(x, y, color='b', linestyle = 'solid', marker = 'o', label = 'apple data')

    plt.xticks(rotation=25)
    plt.xlabel('X Values')
    plt.ylabel('Y Values')
    plt.title('Apple CSV File Information', fontsize=25)
    plt.grid()
    plt.legend()
    plt.show()
```
